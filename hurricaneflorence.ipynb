```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
import matplotlib.pyplot as plt
client = Client(base_url='https://fdsnws.raspberryshakedata.com/')
#client = Client('IRIS')
```

```python
start_1="2018-09-01 00:00:00"
starttime = UTCDateTime(start_1)
end_1="2018-09-16 17:00:00"
endtime = UTCDateTime(end_1)
```

# Durham, NC

```python
waveform1 = client.get_waveforms('AM', 'R3A31', '00', 'SHZ', starttime, endtime)
f = waveform1.filter('bandpass', freqmin=0.025, freqmax=0.25)
f = f.merge()
```

```python
tr = f[0]

p = plt.figure(figsize=(15,5))
ax = p.add_subplot(1,1,1)

ax.set(ylabel='Counts',
       title='AM.R3A31.00.SHZ - Sep 1 00z - Sep 16 17z - 0.025-0.25 Hz')

ax.plot(tr.times("matplotlib"), tr.data, "b-", linewidth=0.5)

ax.set_ylim(-1000,1000)
ax.xaxis_date()
p.autofmt_xdate()
plt.show()
```

```python
waveform1.spectrogram(log=False, title='AM.R3A31 - Sep 1 00z - Sep 16 17z spectrogram')
```
